```python
import xml.etree.ElementTree as ET
import cv2
from statistics import mean
```

```python
def get_element_tree_root(file_name):
    tree = ET.ElementTree(file=file_name)
    root = tree.getroot()
    return root
```

```python
def get_image(page_number):
    image = cv2.imread(".\\images\\page_"+str(page_number)+".jpg")
    return image
```

```python
def draw_outlines(scaling_factor,text_bound,image,color_point):
    x = int(int(text_bound.get('left'))*scaling_factor)
    y = int(int(text_bound.get('top'))*scaling_factor)
    width = int(int(text_bound.get('width'))*scaling_factor)
    height = int(int(text_bound.get('height'))*scaling_factor)+5
    cv2.rectangle(image,(x,y),(x+width,y+height),color_point,2)
```

```python
def save_image(image,page_number):
    cv2.imwrite(".\\output_folder_v5\\editedPage_"+str(page_number)+".jpg",image)
```

```python
def get_color_point(j,holding_list):
    if(j==0):
        if(holding_list[j].get('height') < holding_list[j+1].get('height')):
            if(holding_list[j].text is not None and holding_list[j].text.isnumeric()):
                return False,(0,0,255)
            else:
                return True,(255,0,0)
        else:
            return True,(255,0,0)
    elif(j==len(holding_list)-1):
        if(holding_list[j-1].get('height') > holding_list[j].get('height')):
            if(holding_list[j].text is not None and holding_list[j].text.isnumeric()):
                return False,(0,0,255)
            else:
                return True,(255,0,0)
        else:
            return True,(255,0,0)
    else:
        if(holding_list[j-1].get('height') > holding_list[j].get('height') and holding_list[j].get('height') < holding_list[j+1].get('height')):
            if(holding_list[j].text is not None and holding_list[j].text.isnumeric()):
                return False,(0,0,255)
            else:
                return True,(255,0,0)
        else:
            return True,(255,0,0)
```

```python
def draw_outlines_line(x,y,width,height,scaling_factor,image):
#     print(width)
#     print(height)
    x = int(x*scaling_factor)
    y = int(y*scaling_factor)
    width = int(width*scaling_factor)
    height = int(height*scaling_factor)+5
    cv2.rectangle(image,(x,y),(x+width,y+height),(255,0,0),2)
```

```python
def get_max(parsing_dict):
    min_difference_y = -9999
    for key in parsing_dict.keys():
        if(parsing_dict[key] > min_difference_y):
            min_difference_y = parsing_dict[key]
            index = key
    return index
```
